```python
#1 aggregate > insurance

import os
import json
import pandas as pd

path = r"C:\Users\gopin\OneDrive\Documents\Data_Science\Streamlit Training files\PHONEPE\phpy\Scripts\pulse\data\aggregated\insurance\country\india\state"

agg_ins_state_list = os.listdir(path)


aggr_ins_dic = {'State':[], 'Year':[], 'Quarter':[], 'Transcation_Type':[], 'Transcation_Count':[], 'Transcation_Amount':[]}

for state in agg_ins_state_list:
  agg_ins_year_path = os.path.join(path, state)
  agg_ins_year_list = os.listdir(agg_ins_year_path)

  for year in agg_ins_year_list:
    years_qtr_path = os.path.join(agg_ins_year_path, year)
    years_qtr_list = os.listdir(years_qtr_path)

    for qtr in years_qtr_list:
      qtr_path = os.path.join(years_qtr_path, qtr)
      qtr_file = open(qtr_path, 'r')
      qtr_data = json.load(qtr_file)

      for trans_data in qtr_data['data']['transactionData']:
        aggr_ins_dic['State'].append(state)
        aggr_ins_dic['Year'].append(year)
        aggr_ins_dic['Quarter'].append(int(qtr.strip('.json')))
        aggr_ins_dic['Transcation_Type'].append(trans_data['name'])
        aggr_ins_dic['Transcation_Count'].append(trans_data['paymentInstruments'][0]['count'])
        aggr_ins_dic['Transcation_Amount'].append(trans_data['paymentInstruments'][0]['amount'])


aggr_ins_db = pd.DataFrame(aggr_ins_dic)
aggr_ins_db["State"] = aggr_ins_db["State"].str.replace("-", " ")
aggr_ins_db["State"] = aggr_ins_db["State"].str.title()
aggr_ins_db
```

```python
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://root:Qwerty%40123@localhost/phonepe")

aggr_ins_db.to_sql(
    name="aggregated_insurance",
    con=engine,
    if_exists="replace",
    index=False
)


print("aggregated_insurance created Successfuly")
```

```python
#3 aggregate > User

import os
import json
import pandas as pd

aggr_user_state_path = r"C:\Users\gopin\OneDrive\Documents\Data_Science\Streamlit Training files\PHONEPE\phpy\Scripts\pulse\data\aggregated\user\country\india\state"

aggr_user_state_list = os.listdir(aggr_user_state_path)

aggr_user_dic = {'State':[],
                 'Year':[],
                 'Quarter':[],
                 'Brand':[],
                 'Count':[],
                 'Percentage':[]}


for state in aggr_user_state_list:
  aggr_user_year_path = os.path.join(aggr_user_state_path, state)
  aggr_user_year_list = os.listdir(aggr_user_year_path)

  for year in aggr_user_year_list:
    aggr_user_qtr_path = os.path.join(aggr_user_year_path, year)
    aggr_user_qtr_list = os.listdir(aggr_user_qtr_path)

    for qtr in aggr_user_qtr_list:
      aggr_user_qtr_file_path = os.path.join(aggr_user_qtr_path, qtr)

      with open(aggr_user_qtr_file_path, 'r') as file:
        aggr_user_qtr_data = json.load(file)

      users = aggr_user_qtr_data['data'].get('usersByDevice')

      if users: #Skip Null

        for user_data in users:
            aggr_user_dic['State'].append(state.replace("-", " ").title())
            aggr_user_dic['Year'].append(year)
            aggr_user_dic['Quarter'].append(int(qtr.strip('.json')))
            aggr_user_dic['Brand'].append(user_data['brand'])
            aggr_user_dic['Count'].append(user_data['count'])
            aggr_user_dic['Percentage'].append(user_data['percentage'])


aggr_user_db = pd.DataFrame(aggr_user_dic)
aggr_user_db
```

```python
#6 map > user

import os
import json
import pandas as pd

map_user_state_path = r'C:\Users\gopin\OneDrive\Documents\Data_Science\Streamlit Training files\PHONEPE\phpy\Scripts\pulse\data\map\user\hover\country\india\state'

map_user_state_list = os.listdir(map_user_state_path)


map_user_dic = {'State':[],
                'Year':[],
                'Quarter':[],
                'District_Name':[],
                'Registered_Users':[],
                'App_Opens':[]}


for state in map_user_state_list:
    map_user_year_path = os.path.join(map_user_state_path, state)
    map_user_year_list = os.listdir(map_user_year_path)

    for year in map_user_year_list:
        map_user_qtr_path = os.path.join(map_user_year_path, year)
        map_user_qtr_list = os.listdir(map_user_qtr_path)

        for qtr in map_user_qtr_list:
            map_user_qtr_file_path = os.path.join(map_user_qtr_path, qtr)

            with open(map_user_qtr_file_path, 'r') as file:
                map_user_qtr_data = json.load(file)

                hover_data = map_user_qtr_data.get('data', {}).get('hoverData', {})

                if hover_data:
                    for district_name, values in hover_data.items():
                        map_user_dic['State'].append(state.replace("-", " ").title())
                        map_user_dic['Year'].append(year)
                        map_user_dic['Quarter'].append(int(qtr.strip('.json')))
                        map_user_dic['District_Name'].append(district_name)
                        map_user_dic['Registered_Users'].append(values.get('registeredUsers', 0))
                        map_user_dic['App_Opens'].append(values.get('appOpens', 0))

map_user_db = pd.DataFrame(map_user_dic)
map_user_db
```

```python
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://root:Qwerty%40123@localhost/phonepe")

map_user_db.to_sql(
    name="map_user",
    con=engine,
    if_exists="replace",
    index=False
)

print("map_user created Successfuly")
```
